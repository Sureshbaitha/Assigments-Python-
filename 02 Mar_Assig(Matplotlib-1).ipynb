Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Answer:-Matplotlib is a popular data visualization library in Python used for creating various types of static, interactive, and animated visualizations. It is widely used by data analysts, scientists, and researchers to create high-quality visualizations of their data.

Matplotlib provides several modules for creating different types of visualizations, such as Pyplot, Object-oriented interface, and Animation API. The Pyplot module is the most commonly used module, and it provides a simple interface for creating different types of plots.

Here are five plots that can be plotted using the Pyplot module of Matplotlib:
(a)Line plot - A line plot is used to visualize the relationship between two variables, such as time and temperature.
(b)Bar plot - A bar plot is used to compare different categories of data, such as sales by product category.
(c)Scatter plot - A scatter plot is used to visualize the relationship between two continuous variables, such as height and weight.
(d)Histogram - A histogram is used to visualize the distribution of a single continuous variable, such as the age distribution of a population.
(e)Pie chart - A pie chart is used to show the proportion of different categories in a dataset, such as the percentage of students in different grades.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Answer:-A scatter plot is a type of plot that is used to visualize the relationship between two continuous variables. Each point in the plot represents the values of the two variables for a single observation. Scatter plots are useful for identifying patterns or trends in data.

Here is the code to generate data for x and y and plot a scatter plot using Matplotlib:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title("Scatter Plot of X and Y")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Answer:-The subplot() function is used to create multiple plots within a single figure. It allows us to plot multiple graphs or charts side by side in a single figure, which is useful for comparing and contrasting different datasets or visualizing different aspects of the same data.

The subplot() function takes three arguments: the number of rows, the number of columns, and the plot number. The plot number ranges from 1 to the number of rows times the number of columns, and it specifies which plot to create in the given figure.

Here is the code to create four line plots using the subplot() function with the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2)

# Plot line 1
axes[0, 0].plot(x1, y1)
axes[0, 0].set_title('Line 1')

# Plot line 2
axes[0, 1].plot(x2, y2)
axes[0, 1].set_title('Line 2')

# Plot line 3
axes[1, 0].plot(x3, y3)
axes[1, 0].set_title('Line 3')

# Plot line 4
axes[1, 1].plot(x4, y4)
axes[1, 1].set_title('Line 4')

# Add a common x and y label to the subplots
fig.text(0.5, 0.04, 'X', ha='center')
fig.text(0.04, 0.5, 'Y', va='center', rotation='vertical')

# Display the subplots
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

```python
"""Answer:-A box plot is a graphical representation of a dataset's five-number summary: the minimum, the first quartile, the median, the third quartile, and the maximum. It shows the distribution of a dataset by displaying the range of the data, its central tendency, and how much variation there is in the data. Box plots are used to identify outliers, visualize skewness, and compare distributions between different groups.

To create a box plot using the given data, we first import the required libraries and create two sets of data using the NumPy random.normal() function. Then, we combine these two datasets into a single list and create a box plot using the boxplot() function from the Matplotlib library.

Here is the code to create a box plot from the given data:"""
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine the data
data = [box1, box2]

# Create the box plot
fig, ax = plt.subplots()
ax.boxplot(data)

# Add labels and title
ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')
ax.set_title('Box Plot')

# Show the plot
plt.show()
```
